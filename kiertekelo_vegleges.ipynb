```python
stage_c_f=open("all1.dat","r")
stage_c=stage_c_f.readlines()

cells=[];x1=[];y1=[];x2=[];y2=[]

for j in range(len(stage_c)-1):
    cells.append(float(stage_c[j].split()[0]))
    cells.append(float(stage_c[j].split()[1]))

for i in range(0,len(cells),2):
    x1.append(cells[i])
    y1.append(cells[i+1])

print(len(x1))

x1pure=list(x1);y1pure=list(y1)
print(len(x1pure))
for i in range(len(x1)-1):
    for j in range(len(x1)-1):
        if(abs(x1[j]-x1[i])<=xdlim and abs(y1[j]-y1[i])<=ydlim and i!=j):
            if(x1[i] in x1pure):
                x1pure.remove(x1[i])
            if(y1[i] in y1pure):
                y1pure.remove(y1[i])  

                
print("talált sejtek",len(x1pure))
A=(kep_x_px*resolution)*(kep_y_px*resolution) #mm
print("letett sejtek",str(80*A))
washing_r=(len(x1pure)/A)/80.
print(washing_r)
washing_r=1
```
